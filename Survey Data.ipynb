#### Defining a function that helps split multiple answers into columns

```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
```

```python
lang_work = split_multicolumn(df.LanguageWorkedWith)
lang_work
```

#### LanguageWorkedWith now have a column each

```python
lang_work.sum().sort_values(ascending=False)
```

## Using Ridge

```python
df2 = df[['MainBranch', 'Age', 'Gender', 'Employment', 'EdLevel', 'Age1stCode', 'Hobbyist', 'YearsCode',
         'YearsCodePro', 'CompFreq', 'NEWLearn', 'WorkWeekHrs', 'ConvertedComp', 'JobSat']]
```

```python
df2.dropna(inplace=True)
```

```python
from sklearn.linear_model import Ridge
from category_encoders import OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X = df2[['MainBranch', 'Age', 'Gender', 'Employment', 'EdLevel', 'Age1stCode', 'Hobbyist', 'YearsCode',
         'YearsCodePro', 'CompFreq', 'NEWLearn', 'WorkWeekHrs', 'JobSat']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 40)
```

```python
model = make_pipeline(
    OneHotEncoder(use_cat_names=True),
    Ridge()
)
model.fit(X_train, y_train)
```

```python
prd = model.predict(X_test)
prd
```

```python
model.score(X_test, y_test)
```

```python
me = mean_absolute_error(y_test, prd)
me
```
